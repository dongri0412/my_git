```python
# 导入模块
import pandas as pd
import numpy as np
```

```python
# 导入排列组合计算函数
#调用scipy科学计算包中计算排列组合（permutation and combination）的模块
from scipy.special import perm, comb
#调用 itertools库（内置库） 获取排列组合的全部情况数
from itertools import permutations, combinations
```

```python
path = '大乐透最近100期开奖号码.xlsx'
df = pd.read_excel(path)
df.head()
```

```python
# 红球
df1 = df[:100]
df2 = df1.iloc[:,3:8]
dd = []
for i in range(100):
    dd = dd+list(combinations(list(df2.loc[i]),2))
dd1 = set(dd)

dd2 = []
dd3 = []
dd4 = []
dd5 = []
dd6 = []
a = 0
for i in dd1:
    for ii in dd:
        if i==ii:
            a = a+1
    if a==4:
        dd2.append(i)
    elif a==5:
        dd3.append(i)
    elif a==6:
        dd4.append(i)
    elif a==7:
        dd5.append(i)
    elif a>=8:
        dd6.append(i)       
    a = 0
print(len(dd2),len(dd3),len(dd4),len(dd5),len(dd6))
print(dd2)
print('-'*110)
print(dd3)
print('-'*80)
print(dd4, dd5, dd6)
```

```python
# 篮球
df1 = df[:100]
df2 = df1.iloc[:,8:10]
dd = []
for i in range(100):
    dd = dd+list(combinations(list(df2.loc[i]),2))
dd1 = set(dd)

dd2 = []
dd3 = []
dd4 = []
dd5 = []
dd6 = []
a = 0
for i in dd1:
    for ii in dd:
        if i==ii:
            a = a+1
    if a==4:
        dd2.append(i)
    elif a==5:
        dd3.append(i)
    elif a==6:
        dd4.append(i)
    elif a==7:
        dd5.append(i)
    elif a>=8:
        dd6.append(i)       
    a = 0
print(len(dd2),len(dd3),len(dd4),len(dd5),len(dd6))
print(dd2)
print('-'*110)
print(dd3)
print('-'*80)
print(dd4, dd5, dd6)
```

```python
# 蓝球-单个
df1 = df[:100]
df2 = df1.iloc[:,8:10]
dd = []
for i in range(100):
    dd = dd+list(combinations(list(df2.loc[i]),1))
dd1 = set(dd)

dd2 = []
dd3 = []
dd4 = []
dd5 = []
dd6 = []
dm = []
a = 0
for i in dd1:
    for ii in dd:
        if i==ii:
            a = a+1
    if a>=1 and a<13:
        dd2.append(i)
    elif a>=13 and a<16:
        dd3.append(i)
    elif a>=16 and a<18:
        dd4.append(i)
    elif a>=18 and a<20:
        dd5.append(i)
    elif a>=20:
        dd6.append(i)
    if a>=13:
        dm.append(i)         
    a = 0

    
print(len(dd2),len(dd3),len(dd4),len(dd5),len(dd6))
print(dd2)
print('-'*110)
print(dd3)
print('-'*80)
print(dd4)
print('-'*80)
print(dd5)
print('-'*80)
print(dd6)
print('-'*80)
print("dm=", dm)
print('-'*80)

dxa = []
for i in dm:
    dxa.append(i[0])
dxb = np.sort(dxa)
print(len(dxb), dxb)

```

```python
# 找出dxb与dd7b的交集dm
dm = []
for i in dd7b:
    if i in dxb:
        dm.append(i)
print(len(dm), dm)

df2h = df1.iloc[:,8:10]

dela = []
q =4
for i in range(q):
    for ii in dm:
        if ii in list(df2h.loc[i]):
            dela.append(ii)
print('dm列表中近',q,'期中出现过的蓝球:', dela) 
dmb = []
for i in dm:
    if i not in dela:
        dmb.append(i)

print('dm列表中推荐蓝球:', dmb)
```
