```python
# 爬取大乐透最近100期开奖号码，并且存盘‘大乐透最近100期开奖号码.xlsx’
headers={'user-agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.79 Safari/537.36'}
url='https://webapi.sporttery.cn/gateway/lottery/getHistoryPageListV1.qry?gameNo=85&provinceId=0&pageSize=100&isVerify=1&pageNo=1&termLimits=100'
res = requests.get(url, headers=headers)
print(res)

data = []
d1 = []
for i in res.json()['value']['list']:
    d1.append(int(i['lotteryDrawNum']))
    d1.append(i['lotteryDrawTime'])
    d3 = i['lotteryDrawResult']

    for ii in range(0,20,3):        
        d2 = d3[ii]+d3[ii+1]
        d1.append(int(d2))
    
    d1.append(sum(d1[2:7]))
    d1.append(int(np.mean(d1[2:7])+0.5))
    d1.append(sum(d1[7:9]))
    d1.append(int(np.mean(d1[7:9])+0.5))
    
    ous = 0
    for i in d1[2:7]:
        if i%2 ==0:
            ous = ous+1
    
    d1.append(ous)
    data.append(d1)
    d1 = []

df = pd.DataFrame(data, columns=['期号','日期','H1','H2','H3','H4','H5','L1','L2','红和值','红均值','蓝和值','蓝均值','红偶数个数'])

df['年份'] = 0
df['年内期号'] = 0
df['年内月号'] = 0
for i in range(len(df.期号)):
    df['年份'][i] = str(df.日期[i])[0:4]
    df['年内期号'][i] = str(df.期号[i])[-3:]
    df['年内月号'][i] = str(df.日期[i])[5:7]

df.to_excel('大乐透最近100期开奖号码.xlsx', index=True)
# columns= Unnamed: 0	期号	日期	H1	H2	H3	H4	H5	L1	L2	红和值	红均值	蓝和值	蓝均值	红偶数个数	年份	年内期号	年内月号
df.head()

```

## 计算：12个月份各月的红球分布数据

```python
# 计算：12个月份各月的红球分布数据
yy = []
for i in range(12):
    d = df3_2[i][1]
    y = []
    for a in range(1,37):
        y.append(a)
    y[0] = i+1
    x = list(range(0,36))

    for ii in range(1,36):
        for h in d.H1:
            if h==ii:
                y[ii] = y[ii]+1
        for h in d.H2:
            if h==ii:
                y[ii] = y[ii]+1
        for h in d.H3:
            if h==ii:
                y[ii] = y[ii]+1
        for h in d.H4:
            if h==ii:
                y[ii] = y[ii]+1
        for h in d.H5:
            if h==ii:
                y[ii] = y[ii]+1
    
    yy.append(y)
    
    # 创建画布 p，设置 x_axis_type='datetime'
    # plt.figure(figsize=(12,6),dpi=400)
    #plt.figure()
    #plt.plot(x, y)
    #plt.show()
    #plt.bar(x, y)
    #plt.title='月份'
    #plt.show()
    
dyy = pd.DataFrame(yy,columns=x)
dyy.to_excel('12个月份各月的红球分布数据.xlsx', index=False)
xy = pd.read_excel('12个月份各月的红球分布数据.xlsx')
xy
```

## 画出各月份中每种红球的出击次数大小（即：色点大小表示出击次数多少）

```python
# 导入需要的函数
from bokeh.plotting import figure, output_notebook, show
# import pandas as pd
# import numpy as np
# import matplotlib.pyplot as plt
# import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# 画出各月份中每种红球的出击次数大小（即：色点大小表示出击次数多少）
y_list = []
y1 = []
for b1 in range(1,36):
    for b2 in range(12):
        y1.append(b1)
    y_list.append(y1)
    y1 = []

# 创建画布 p，设置 x_axis_type='datetime'
# plt.figure(x_axis_label='年内月号', y_axis_label='出球编号')
plt.figure(figsize=(10,6),dpi=600)
plt.xlabel('x = YUEFEN')
plt.ylabel('y = HONGQIIU_NUM')
plt.title('DLT_HONGQIU_NUM_SIZE')
plt.grid(True)#显示网格


for i in range(35):
    y = y_list[i]
    y2 = xy[i+1]
    xy_mean = xy[i+1].mean()
    for c in range(len(y2)):
        if y2[c]>xy_mean:
            y2[c]=y2[c]*2*(y2[c]/xy_mean)
        else:
            y2[c]=y2[c]*0.8*(y2[c]/xy_mean)
        
    x = xy[0]

    # 在x 轴绘制'年内月号'，沿 y 轴绘制出球编号,散点图
    # plt.scatter(x, y, s=y2, label=y[0])  #显示图例
    plt.scatter(x, y, s=y2)  #无图例
# plt.legend(loc='best') # 显示图例时候采用该句代码
plt.show()

xy = pd.read_excel('12个月份各月的红球分布数据.xlsx')
```

```python
# 计算：12个月份各月的蓝球分布数据
yy = []
for i in range(12):
    d = df3_2[i][1]
    y = []
    for a in range(13):
        y.append(a)
    y[0] = i+1
    x = list(range(13))

    for ii in range(1,13):
        for L in d.L1:
            if L==ii:
                y[ii] = y[ii]+1
        for L in d.L2:
            if L==ii:
                y[ii] = y[ii]+1
            
    yy.append(y)
dyyl = pd.DataFrame(yy,columns=x)
dyyl.to_excel('12个月份各月的蓝球分布数据.xlsx', index=False)
xyl = pd.read_excel('12个月份各月的蓝球分布数据.xlsx')
xyl
```
