```python
# 取当前日期和时间
import time
date = (time.strftime("%Y-%m-%d %H:%M:%S", time.localtime()))
print(date)
```

```python
# 导入模块
import pandas as pd
import numpy as np
```

```python
# 导入排列组合计算函数
#调用scipy科学计算包中计算排列组合（permutation and combination）的模块
from scipy.special import perm, comb
#调用 itertools库（内置库） 获取排列组合的全部情况数
from itertools import permutations, combinations
```

```python
path = '大乐透总历史开奖号码表.xlsx'
df = pd.read_excel(path)
print(df.shape)
df.head()
```

```python
# 蓝球-单个
print("## 蓝球-单个:")
df1 = df[:100]
df2 = df1.iloc[:,7:9]
dd = []
for i in range(100):
    dd = dd+list(combinations(list(df2.loc[i]),1))
    #print(df2.loc[i])
dd1 = set(dd)

dd2 = []
dd3 = []
dd4 = []
dd5 = []
dd6 = []
dm = []
a = 0
for i in dd1:
    for ii in dd:
        if i==ii:
            a = a+1
    if a>=1 and a<13:
        dd2.append(i)
    elif a>=13 and a<15:
        dd3.append(i)
    elif a>=15 and a<18:
        dd4.append(i)
    elif a>=18 and a<23:
        dd5.append(i)
    elif a>=23:
        dd6.append(i)
    a = 0

dm = dd2+dd6    
print('#### 各组球数分别为：', len(dd2),len(dd3),len(dd4),len(dd5),len(dd6))
print('#### a<13:           ', dd2)
print('-'*80)
print('#### a>=13 and a<15: ', dd3)
print('-'*80)
print('#### a>=15 and a<18: ', dd4)
print('-'*80)
print('#### a>=18 and a<23: ', dd5)
print('-'*80)
print('#### a>23:           ',  dd6)
print('-'*80)


dxa = []
for i in dm:
    dxa.append(i[0])
dxb = np.sort(dxa)
print("### 100期选取区间a<13ora>=23:", len(dxb), '个，', dxb)
print('='*100)


# 篮球-单个
df1 = df[:25]
df2 = df1.iloc[:,7:9]
dd = []
for i in range(25):
    dd = dd+list(combinations(list(df2.loc[i]),1))
dd1 = set(dd)

dd2 = []
dd3 = []
dd4 = []
dd5 = []
dd6 = []
a = 0
for i in dd1:
    for ii in dd:
        if i==ii:
            a = a+1
    if a>=1 and a<2:
        dd2.append(i)
    elif a>=2 and a<3:
        dd3.append(i)
    elif a>=3 and a<6:
        dd4.append(i)
    elif a>=6 and a<13:
        dd5.append(i)
    elif a>=13:
        dd6.append(i)
    if a>=13:
        dm.append(i)         
    a = 0
    
print('#### 各组球数分别为：', len(dd2),len(dd3),len(dd4),len(dd5),len(dd6))
print('#### a=1:          ', dd2)
print('-'*80)
print('#### a=2:          ', dd3)
print('-'*80)
print('#### a=3~5:        ', dd4)
print('-'*80)
print('#### a=6~13:        ', dd5)
print('-'*80)
print('#### a>=13:        ', dd6)
print('-'*80)

dd7 = dd4+dd3
dd7a = []
for i in dd7:
    dd7a.append(i[0])
dd7b = list(np.sort(dd7a))
print("### 25期热度区间a=2~5: ", len(dd7b), '个，',dd7b)
print('='*100)


# 找出100期出击最多区间dmax与25期热度区间dd7b的交集dm
dm = []
for i in dd7b:
    if i in dxb:
        dm.append(i)
print('#### 找出100期选取区间a<13ora>=23与25期热度区间a=2~5的交集dm: ', len(dm), '个，', dm)
 
df2h = df1.iloc[:,7:9]

dela1 = []
q1 =3
for i in range(q1):
    for ii in dm:
        if ii in list(df2h.loc[i]):
            dela1.append(ii)
print('#### dm列表中近',q1,'期中出现过的蓝球:', dela1) 
dmc = []
for i in dm:
    if i not in dela1:
        dmc.append(i)

print('### 篮球列表中推荐蓝球:', len(dmc), '个，', dmc)

```

```python
# 红球-单个，A组：
print("## 红球-单个，A组：")
df1 = df[:100]
df2 = df1.iloc[:,2:7]
dd = []
for i in range(100):
    dd = dd+list(combinations(list(df2.loc[i]),1))
dd1 = set(dd)

dd2 = []
dd3 = []
dd4 = []
dd5 = []
dd6 = []
a = 0
for i in dd1:
    for ii in dd:
        if i==ii:
            a = a+1
    if a>=1 and a<10:
        dd2.append(i)
    elif a>=10 and a<13:
        dd3.append(i)
    elif a>=13 and a<15:
        dd4.append(i)
    elif a>=15 and a<17:
        dd5.append(i)
    elif a>=17:
        dd6.append(i)       
    a = 0

dmax = dd2+dd6 
print('#### 各组球数分别为：',len(dd2),len(dd3),len(dd4),len(dd5),len(dd6))
print('#### a<10:          ', dd2)
print('-'*80)
print('#### a>=10 and a<13:', dd3)
print('-'*80)
print('#### a>=13 and a<15:', dd4)
print('-'*80)
print('#### a>=15 and a<17:', dd5)
print('-'*80)
print('#### a>=17:', dd6)
print('-'*80)


dxa = []        
for i in dmax:
    dxa.append(i[0])
dxb = list(np.sort(dxa))
print("### dmax a<10 or a>=17: ", len(dxb), '个，',dxb)
print('='*100)

# 红球-单个
df1 = df[:25]
df2 = df1.iloc[:,2:7] 
dd = []
for i in range(25):
    dd = dd+list(combinations(list(df2.loc[i]),1))
dd1 = set(dd)

dd2 = []
dd3 = []
dd4 = []
dd5 = []
dd6 = []
a = 0
for i in dd1:
    for ii in dd:
        if i==ii:
            a = a+1
    if a>=1 and a<3:
        dd2.append(i)
    elif a==3:
        dd3.append(i)
    elif a==4:
        dd4.append(i)
    elif a==5:
        dd5.append(i)
    elif a>=6:
        dd6.append(i)       
    a = 0
print('#### 各组球数分别为：',len(dd2),len(dd3),len(dd4),len(dd5),len(dd6))
print('#### a<3:', dd2)
print('-'*80)
print('#### a=3:', dd3)
print('-'*80)
print('#### a=4:', dd4)
print('-'*80)
print('#### a=5:', dd5)
print('-'*80)
print('#### a>6:', dd6)
print('-'*80)

dd7 = dd4+dd5+dd6
dd7a = []
for i in dd7:
    dd7a.append(i[0])
dd7b = list(np.sort(dd7a))
print("#### 热度区间 a>=4: ", len(dd7b), dd7b)
print('='*100)

# 找出100期出击最多区间dmax与25期热度区间dd7b的交集dm
dm = []
for i in dd7b:
    if i in dxb:
        dm.append(i)
print("#### 找出100期选取区间dmax与25期热度区间的交集：", len(dm), '个，', dm)

df2h = df1.iloc[:,2:7]

dela2 = []
q2 =7
for i in range(q2):
    for ii in dm:
        if ii in list(df2h.loc[i]):
            dela2.append(ii)
print('#### dm列表中近', q2, '期中出现过的红球:', dela2, '=>', set(dela2)) 
dma = []
for i in dm:
    if i not in dela2:
        dma.append(i)

print('### 红球A组列表中余下的推荐红球:',  len(dma), '个，',dma)
print('### 红球A组列表中推荐近{}期热度红球: {}'.format(q2, set(dela2)))
```

```python
# 红球-单个,B组：
print("## 红球-单个，B组：")
df1 = df[:100]
df2 = df1.iloc[:,2:7]
dd = []
for i in range(100):
    dd = dd+list(combinations(list(df2.loc[i]),1))
dd1 = set(dd)

dd2 = []
dd3 = []
dd4 = []
dd5 = []
dd6 = []
a = 0
for i in dd1:
    for ii in dd:
        if i==ii:
            a = a+1
    if a>=1 and a<10:
        dd2.append(i)
    elif a>=10 and a<13:
        dd3.append(i)
    elif a>=13 and a<15:
        dd4.append(i)
    elif a>=15 and a<17:
        dd5.append(i)
    elif a>=17:
        dd6.append(i)       
    a = 0

dmax = dd2+dd6 
print('#### 各组球数分别为：', len(dd2),len(dd3),len(dd4),len(dd5),len(dd6))
print('#### a<10:          ', dd2)
print('-'*80)
print('#### a>=10 and a<13:', dd3)
print('-'*80)
print('#### a>=13 and a<15:', dd4)
print('-'*80)
print('#### a>=15 and a<17:', dd5)
print('-'*80)
print('#### a>=17:', dd6)
print('-'*80)


dxa = []        
for i in dmax:
    dxa.append(i[0])
dxb = list(np.sort(dxa))
# print("### dmax a>=17: ", len(dxb), dxb)
print("### dmax a<10 or a>=17: ", len(dxb), '个，',dxb)
print('='*100)

# 红球-单个
df1 = df[:25]
df2 = df1.iloc[:,2:7] 
dd = []
for i in range(25):
    dd = dd+list(combinations(list(df2.loc[i]),1))
dd1 = set(dd)

dd2 = []
dd3 = []
dd4 = []
dd5 = []
dd6 = []
a = 0
for i in dd1:
    for ii in dd:
        if i==ii:
            a = a+1
    if a>=1 and a<3:
        dd2.append(i)
    elif a==3:
        dd3.append(i)
    elif a==4:
        dd4.append(i)
    elif a==5:
        dd5.append(i)
    elif a>=6:
        dd6.append(i)       
    a = 0
print('#### 各组球数分别为：',len(dd2),len(dd3),len(dd4),len(dd5),len(dd6))
print('#### a<3:', dd2)
print('-'*80)
print('#### a=3:', dd3)
print('-'*80)
print('#### a=4:', dd4)
print('-'*80)
print('#### a=5:', dd5)
print('-'*80)
print('#### a>6:', dd6)
print('-'*80)

dd7 = dd3+dd4
dd7a = []
for i in dd7:
    dd7a.append(i[0])
dd7b = list(np.sort(dd7a))
print("### 热度区间 a=3 and a=4: ", len(dd7b), dd7b)
print('='*100)

# 找出dxb=dmax与dd7b的交集dm
dm = []
for i in dd7b:
    if i in dxb:
        dm.append(i)
print("#### 找出100期选取区间dmax与25期热度区间的交集dm：", len(dm), '个，', dm)

df2h = df1.iloc[:,2:7]

dela3 = []
dela3a = []
q3 =7
q3a = 4
dma = []
for i in range(q3):
    for ii in dm:
        if ii in list(df2h.loc[i]):
            dela3.append(ii)
print('#### dm列表中近',q3,'期中出现过的红球dma:', dela3) 
dma = set(dela3)

for i in range(q3a):
    for ii in dma:
        if ii not in list(df2h.loc[i]):
            dela3a.append(ii)

print('#### dm列表中近{}期出过红球dma组中，推荐近{}期未出的红球{}:'.format(q3, q3a, set(dela3a)) )

dmb = []
for i in dm:
    if i not in dela3:
        dmb.append(i)

print('### 红球B组列表中余下推荐红球:', len(dmb), '个，', dmb)

```
